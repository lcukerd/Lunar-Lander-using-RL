```python
import numpy as np
from random import randint
import gym
```

```python
def createFirstGeneration(population,length):
    individuals = np.random.randint(4,size = length)
    for i in range(population-1):
        individuals = np.vstack((individuals,np.random.randint(4,size = length)))
    return individuals
```

```python
def createChild(parent1,parent2):
    child = np.array([])
    for i in range(parent1.shape[0]):
        if (randint(0,1) == 1):
            child = np.append(child,parent1[i])
        else:
            child = np.append(child,parent2[i])
    return child
```

```python
def mutation(individual):
    for i in range(individual.shape[0]):
        if (randint(0,6) == 5):
            individual[i,randint(0,individual.shape[1]-1)] = randint(0,3)
    return individual
```

```python
def Selection(top,lucky,population,fitness):
    inds = fitness.argsort()
    population = population[inds]
    Selected = np.array(population[population.shape[0]-1])
    for i in range(population.shape[0]):
        if (i >= population.shape[0] - top):
            Selected = np.vstack((Selected,population[i]))
        elif (randint(1,10) == 5 & lucky >0):
            Selected = np.vstack((Selected,population[i]))
            lucky -=1
    return Selected
```

```python
def breeding(Population,childs):
    NextGen = np.array(createChild(Population[0],Population[1]))
    for i in range(0,Population.shape[0]-1):
        for j in range(childs):
            NextGen = np.vstack((NextGen,createChild(Population[i],Population[i+1])))
    return NextGen
```

```python
def Fitness(population,env):
    fitness = np.array([])
    for i in range(population.shape[0]):
        j = 0
        total_reward = 0
        env.reset()
        done = False
        while not done:
            _,reward,done,_ = env.step(int(population[i,j]))
            total_reward += reward
            j += 1
        fitness = np.append(fitness,total_reward)
    return fitness
```

```python
def play(individual,env):
    total_reward = 0
    env.reset()
    j = 0
    done = False
    while not done:
        env.render()
        _,reward,done,_ = env.step(int(individual[j]))
        total_reward += reward
        j += 1
    print (total_reward,end = ' , ')
```

```python
env = gym.make('LunarLander-v2')
Individuals = createFirstGeneration(1000,1000)
fitness = Fitness(Individuals,env)
Generations = 5000
for j in range(Individuals.shape[0]):
    print (str(fitness[j]),end = ' , ')
```

```python
for i in range(Generations):
    print ('\n Generations : ',i, ' Population : ',Individuals.shape[0])
    Individuals = Selection(100,10,Individuals, fitness)
    Individuals = breeding(Individuals,10)
    Individuals = mutation(Individuals)
    fitness = Fitness(Individuals,env)
    
    inds = (fitness).argsort()
    Individuals = Individuals[inds]
    fitness = fitness[inds]
    
    for j in range(Individuals.shape[0]-10,Individuals.shape[0]):
        play(Individuals[i],env)
env.close()
```
